```python
# Import essential libraries
import numpy as np
import pandas as pd
import seaborn as sb

# import the Random Under Sampler object.
from imblearn.under_sampling import RandomUnderSampler
```

```python
#import the dataset
diabetes = pd.read_csv('dataset/diabetes_binary.csv')
diabetes
```

```python
sb.catplot(y='Diabetes_binary', data = diabetes, kind = "count")
```

<br>

After plotting the histogram, it is obvious that it is an imbalanced dataset as there are too many people do not have diabetes in the dataset (negative instance).

This will lead to **poor model performance** on the dataset and **high error rate for minority class** (positive instance). Besides, this will lead to **misleading performance metrics**, accuracy can be misleading on imbalanced dataset as it will predict all negative instances correctly.

In order to handle imbalance dataset, we are going to use the technique of **Undersampling**. **Undersampling** is a way to balance our dataset by randomly removing some data from the larger group (negative instances) until it matches the size of the smaller group (positive instances).

We are going to use the function of `RandomUnderSampler()` to achieve this technique. Finally, we concatenate two resampled dataframe and plot the histogram. It is obvious that the dataset is balanced.

<br>

```python
y = pd.DataFrame(diabetes['Diabetes_binary'])
x = diabetes.drop('Diabetes_binary', axis=1)

undersampler = RandomUnderSampler()
x_resampled, y_resampled = undersampler.fit_resample(x,y)
resampled = pd.concat([y_resampled, x_resampled], axis=1)
```

```python
sb.catplot(y='Diabetes_binary', data = resampled, kind = "count")
```

We are exploring our data used by knowing the datatype and the statistical information of each column. The method `.info()` and `.describe()` are used.

```python
resampled.info()
```

```python
#understanding the basic statistical information for each column
resampled.describe()
```

<br>

As most of our variables are categorical data and the data in each variables are mostly `float`. Thus, we firstly sort out the non-categorical data (since the majority of the data are categorical), then the rest we change hte data type of it into `integer` to facilitate our further process or comparison.

<br>

```python
#list out the non-categorical data
list = ['Diabetes_binary', 'BMI']

#change the datatype of the categorical data
for i in resampled:
    if (i not in list):
        resampled[i] = resampled[i].astype('int64')
resampled.info()
```

<br>

Lastly, we export the organized data into a new file (**resampled.csv**) for out further use.

<br>

```python
#export organized data into a new file
file_path = "dataset/resampled.csv"

resampled.to_csv(path_or_buf = file_path, index=False)
```
